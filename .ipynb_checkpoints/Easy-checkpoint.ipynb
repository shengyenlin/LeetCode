# 7. Reverse Integer
- Given a signed 32-bit integer x, return x with its digits reversed. 
- If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

```python
class Solution:
    def reverse(self, x: int) -> int:
        isNeg = (x < 0)
        lower_bound = (-2) ** 31
        upper_bound = (2) ** 31 - 1
        str_x = str(x)
        if isNeg:
            str_x = str_x[1:]
            
        len_x = len(str_x)
        str_out = ""
        for idx, el in enumerate(str_x):
            str_out += str_x[len_x - idx - 1]
#             print(idx, str_x[len_x - idx - 1], str_out)
        
        int_out = int(str_out)
        isGreaterThanUb = (int_out > upper_bound)
        isLowerThanLb = (int_out < lower_bound)
        if isGreaterThanUb or isLowerThanLb:
            return 0
        elif isNeg:
            return (-int_out)
        else:
            return int_out
```

## My Solution
- Time Complexity : O(N), N = length of x 
- Space Complexity: O(1)



```python
sol = Solution()
sol.reverse(0)
```

## Sample Solution
- Time Complexity: O(log(N))
- Space Complexity: O(1)

```python
class Solution:
    def reverse(self, x: int) -> int:
        lower_bound = (-2) ** 31
        upper_bound = (2) ** 31 - 1
        isNeg = (x < 0)

        if x == 0:
            return 0
        if isNeg:
            x = -x
        
        if (x < 10):
            ans = x

        ans = 0
        while (x != 0):
            #pop the first digit
            pop = int(x % 10)
            x = x // 10
            #push back
            
            temp = ans * 10 + pop
            isGreaterThanUb = (temp > upper_bound)
            isLowerThanLb = (temp < lower_bound)
            if isGreaterThanUb or isLowerThanLb:
                return 0
            else:
                ans = temp
                
            #print(pop, x, ans, "\n")
        
        if isNeg:
            return -ans
        else:
            return ans
```

```python
sol = Solution()
sol.reverse(120)
```
